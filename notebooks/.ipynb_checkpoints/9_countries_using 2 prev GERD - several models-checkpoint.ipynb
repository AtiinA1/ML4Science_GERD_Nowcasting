```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.callbacks import LearningRateScheduler
from sklearn.metrics import mean_absolute_error
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings('ignore')
```

```python
iso = ['CAN', 'CHN', 'FRA', 'DEU', 'JPN', 'KOR', 'CHE', 'GBR', 'USA']

df = pd.read_csv('../data/DP_LIVE_08052023154811337.csv')
df_filtered = df[(df['MEASURE']=='PC_GDP') &
                (df['TIME']>=2008)&
                (df['TIME']<=2020)]
df_filtered = df_filtered[df_filtered['LOCATION'].isin(iso)][['LOCATION','TIME', 'Value']]
display(df_filtered)
```

```python

df = df_filtered


df = df.sort_values(by=['LOCATION', 'TIME'])

# Then, we will shift the 'Value' column 
df['v-1'] = df.groupby('LOCATION')['Value'].shift(1)
df['v-2'] = df.groupby('LOCATION')['Value'].shift(2)



# drop rows that do not have complete data for the previous two years.
df = df.dropna(subset=['v-1', 'v-2']).reset_index(drop=True)


df.rename(columns={'LOCATION': 'location', 'TIME': 'time', 'Value': 'v'}, inplace=True)

df = df[['location', 'time', 'v-2', 'v-1', 'v']]

df 
```

```python


#split data
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# normalize train and test data

mean_train = train_df[['v-2', 'v-1']].mean()
std_train = train_df[['v-2', 'v-1']].std()


train_df[['v-2', 'v-1']] = (train_df[[ 'v-2', 'v-1']] - mean_train) / std_train


test_df[[ 'v-2', 'v-1']] = (test_df[[ 'v-2', 'v-1']] - mean_train) / std_train



# seperate features and outcome
X_train = train_df.drop(['v','location', 'time'], axis=1)
y_train = train_df['v']


X_test = test_df.drop(['v','location', 'time'], axis=1)
y_test = test_df['v']
```

### baseline: linear regression, ARIMA



### Linear regression


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

#linear regression model

model = LinearRegression()

model.fit(X_train, y_train)

y_pred_train = model.predict(X_train)

y_pred_test = model.predict(X_test)


#evaluate the model
train_performance = model.score(X_train, y_train)
test_performance = model.score(X_test, y_test)


print("Performance on Training Data (R² score)  ", train_performance)
print("Performance on Test Data (R² score)  ", test_performance)


#test loss (MSE)
test_loss = mean_squared_error(y_test, y_pred_test)
print("Test Loss (Mean Squared Error)", test_loss)

#test loss (MAE)
mae = mean_absolute_error(y_test, y_pred_test)
print(f"Mean Absolute Error on Test Set: {mae}")
```

```python
# MAPE
np.mean(np.abs((y_pred_test.reshape(-1) - y_test.values.reshape(-1)) /y_test.values.reshape(-1)))*100
```

### ARIMA

```python

model = ARIMA(y_train, order=(1, 1, 1))
model_fit = model.fit()

# Forecast
y_pred_test = model_fit.forecast(steps=len(y_test))

# Evaluate the model
test_loss = mean_squared_error(y_test, y_pred_test)
print("Test Loss (Mean Squared Error)", test_loss)
```

```python
mae = mean_absolute_error(y_test, y_pred_test)
print(f"Mean Absolute Error on Test Set: {mae}")

# MAPE
np.mean(np.abs((y_pred_test.values.reshape(-1) - y_test.values.reshape(-1)) /y_test.values.reshape(-1)))*100
```

### Neural Network

##### MLP

```python


#  model: nn with 2 fully connected layers
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1)  
])


model.compile(optimizer='adam', loss='mean_squared_error')


history = model.fit(X_train, y_train, epochs=1000, validation_split=0.1)  

# evaluate the model
test_loss = model.evaluate(X_test, y_test)
print(f"Test Loss: {test_loss}")

# prediction
y_pred = model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error on Test Set: {mae}")
# MAPE
np.mean(np.abs((y_pred.reshape(-1) - y_test.values.reshape(-1)) /y_test.values.reshape(-1)))*100
```

##### several layers

```python


# learning rate scheduler
def scheduler(epoch, lr):
    if epoch < 20:
        return lr
    else:
        return lr * tf.math.exp(-0.1)

# neural network model
model = Sequential([
    Dense(1024, input_shape=(2,)),
    Activation('relu'),
    Dense(512),
    Activation('relu'),
    Dense(256),
    Activation('relu'),
    Dense(128),
    Activation('relu'),
    Dense(64),
    Activation('relu'),
    Dense(32),
    Activation('relu'),
    Dense(1)
])


initial_learning_rate = 0.01
optimizer = tf.keras.optimizers.Adam(learning_rate=initial_learning_rate)
model.compile(optimizer=optimizer, loss='mean_squared_error')

#Learning rate scheduler callback
lr_scheduler = LearningRateScheduler(scheduler)

# Fit the model
model.fit(X_train, y_train, epochs=150, batch_size=10, callbacks=[lr_scheduler], validation_split=0.2)

# Prediction
y_pred = model.predict(X_test)

#MAE loss
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error on Test Set: {mae}")
```

```python
# MAPE
np.mean(np.abs((y_pred.reshape(-1) - y_test.values.reshape(-1)) /y_test.values.reshape(-1)))*100
```

### KNN

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt

X = df.drop(['v', 'location'], axis=1).values  
y = df.drop('location', axis=1)['v'].values  

k_values = range(1, 10)

average_mse_scores = []

for k in k_values:
    knn_regressor = KNeighborsRegressor(n_neighbors=k)
    scores = cross_val_score(knn_regressor, X, y, cv=5, scoring='neg_mean_squared_error')
    average_mse_scores.append(np.mean(-scores))

plt.figure(figsize=(10, 6))
plt.plot(k_values, average_mse_scores, marker='o')
plt.title('KNN Regression: MSE for Different Values of k')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Average MSE')
plt.xticks(k_values)
plt.show()
```

```python
# knn
knn_regressor = KNeighborsRegressor(n_neighbors=4)

# Cross-Validation
scores = cross_val_score(knn_regressor, X, y, cv=5, scoring='neg_mean_squared_error')

# Calculate average MSE
average_mse = np.mean(-scores)
print(f"Average MSE: {average_mse}")
```

KNN is not performing well. But it could jave been a good idea if it performed wel as it would possible to use it as a model to estimate GERD on finer granularity
