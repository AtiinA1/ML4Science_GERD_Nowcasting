##  Gather Guerd Data

```python

iso = ['CAN', 'CHN', 'FRA', 'DEU', 'JPN', 'KOR', 'CHE', 'GBR', 'USA']

def load_R_D_expenditure_data(iso):
    """
    Load and preprocess R&D expenditure data.

    This function reads a CSV file containing R&D expenditure data, filters and preprocesses it to create a DataFrame
    suitable for analysis. It also calculates the difference between consecutive R&D expenditure values for future use.

    Parameters:
        iso (list of str): List of ISO country codes to filter data by.

    Returns:
        pandas.DataFrame: A DataFrame containing the following columns:
            - 'ISO': ISO country code.
            - 'year': Year of the data.
            - 'v-1': R&D expenditure for the previous year.
            - 'v-2': R&D expenditure for the year before the previous year.
            - 'v': R&D expenditure for the current year.
    """
    
    # Loading the dataframe with all the R&D expenditure
    df = pd.read_csv('./data/DP_LIVE_08052023154811337.csv')
    
    # Filter on the measure, time, and location
    df_GERD = df[(df['MEASURE']=='PC_GDP') &
                 (df['TIME'] >= 2003)]
    df_GERD = df_GERD[df_GERD['LOCATION'].isin(iso)][['LOCATION','TIME', 'Value']]
    df_GERD = df_GERD.sort_values(by=['LOCATION', 'TIME'])
    
    # Then, we will shift Our window in the 'Value' column, we use this since we want to predict R&D expenditures with previous 
    # values as features
    df_GERD['v-1'] = df_GERD.groupby('LOCATION')['Value'].shift(1)
    df_GERD['v-2'] = df_GERD.groupby('LOCATION')['Value'].shift(2)

    # Drop rows that do not have complete data for the previous two years.
    df_GERD = df_GERD.dropna(subset=['v-1', 'v-2']).reset_index(drop=True)

    df_GERD.rename(columns={'LOCATION': 'ISO', 'TIME': 'year', 'Value': 'v'}, inplace=True)
    df_GERD = df_GERD[['ISO', 'year', 'v-1', 'v-2', 'v']]
    
    # Here we adopt the approach of predicting the difference between 2 consecutive values of R&D expenditures rather than predicting
    # the actual value, this will be useful next when we want to predict R&D expenditures on a quarterly basis.
    
    return df_GERD
```

```python
def reshape_dataframe(df, name):
    """
    Reshape a given DataFrame to a long format using the pandas melt function.

    This function is specifically designed to reshape data related to yearly 
    information, transforming it from a wide format (where each year is a separate 
    column) to a long format (where years are rows under a single column).

    Parameters:
    df (pandas.DataFrame): The DataFrame to be reshaped.
    name (str): The name to be given to the value column after reshaping.

    Returns:
    pandas.DataFrame: The reshaped DataFrame in long format.
    """

    gdp_reshaped = df.melt(id_vars=['ISO'], 
                           value_vars=[str(year) for year in range(2010, 2023)],
                           var_name='year', 
                           value_name=name)

    return gdp_reshaped
```

```python
# Reshape each dataframe
gdp_reshaped = reshape_dataframe(gdp_df, 'GDP')
inflation_reshaped = reshape_dataframe(inflation_df, 'Inflation')
govt_debt_reshaped = reshape_dataframe(govt_debt_df,  'Govt_Debt')
unemployment_reshaped = reshape_dataframe(unemployment_df, 'Unemployment')
govt_expenditure_reshaped = reshape_dataframe(gvt_expenditure, 'Govt_Expenditure')
total_investment_reshaped = reshape_dataframe(total_inv_df, 'Total_Investment')
national_savings_reshaped = reshape_dataframe(gvt_savings,  'National_Savings')
inflation_eop_reshaped = reshape_dataframe(inflation_eop_df,  'Inflation_EoP')
current_account_balance_reshaped = reshape_dataframe(balance_df,  'Account_Balance')

# Combine all reshaped dataframes using inner merge
from functools import reduce
data_frames = [gdp_reshaped, inflation_reshaped, govt_debt_reshaped, unemployment_reshaped, govt_expenditure_reshaped, total_investment_reshaped, national_savings_reshaped, inflation_eop_reshaped, current_account_balance_reshaped]
combined_df_macro_vars = reduce(lambda left, right: pd.merge(left, right, on=['ISO', 'year'], how='inner'), data_frames)

## shift combined df to get for each year values from the previous yeear
combined_df_macro_vars['year'] = combined_df_macro_vars['year'].astype('int')+ 1

# Display the first few rows of the combined dataframe
combined_df_macro_vars.head(3)
```

```python
#Merge GERD and macro variables data and process macro variables accordingly
GERD_and_macro_vars = pd.merge(merged_df_R_D, combined_df_macro_vars, on=['ISO', 'year'], how='inner')
GERD_and_macro_vars_data = GERD_and_macro_vars.drop(axis=1,columns=["ISO","year"])
GERD_and_macro_vars_data = GERD_and_macro_vars_data.fillna(0)
GERD_and_macro_vars_data['GDP'] = GERD_and_macro_vars_data['GDP'].str.replace(',', '', regex=False)
columns_to_convert = ['GDP', 'Inflation', 'Govt_Debt', 'Unemployment', 'Govt_Expenditure', 'Total_Investment', 'National_Savings', 'Inflation_EoP', 'Account_Balance']

for col in columns_to_convert:
    GERD_and_macro_vars_data[col] = pd.to_numeric(GERD_and_macro_vars_data[col], errors='coerce')
GERD_and_macro_vars_data
```

## Define models

```python
def MAPE(y_test,y_pred_test) : 
    return np.mean(np.abs(y_test - y_pred_test)/y_test*100)
```

```python
def Linear_regression_model(X_train,y_train,X_test,y_test) :
    
    #linear regression model
    model = LinearRegression()
    model.fit(X_train, y_train)
    y_pred_test = model.predict(X_test)
    
    #test loss (MAE)
    mae = mean_absolute_error(y_test, y_pred_test)
    #test loss (MAPE)
    mape = MAPE(y_test,y_pred_test)

    
    return mae,mape
    
```

```python
def ARIMA_model(X_train,y_train,X_test,y_test) :
    
    model = ARIMA(y_train, order=(1, 1, 1))
    model_fit = model.fit()

    # Forecast
    y_pred_test = model_fit.forecast(steps=len(y_test))

    #test loss (MAE)
    mae = mean_absolute_error(y_test, y_pred_test)
    #test loss (MAPE)
    mape = MAPE(y_test,y_pred_test)
    
    return mae,mape
```

```python
def NN_1layer_model(X_train,y_train,X_test,y_test) :
    #  model: nn with 2 fully connected layers
    model = Sequential([
        Dense(2048, activation='relu', input_shape=(X_train.shape[1],)),
        Dense(1)  
    ])


    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(X_train, y_train, epochs=120, validation_split=0.1)  
    y_pred_test = model.predict(X_test)
    
    #test loss (MAE)
    mae = mean_absolute_error(y_test, y_pred_test.reshape(-1))
    #test loss (MAPE)
    mape = MAPE(y_test,y_pred_test.reshape(-1))
    
    return mae,mape
```

```python
def LSTM_model(X_train,y_train,X_test,y_test) :
    
    lstm_x_train = X_train[[ 'v-2', 'v-1']].to_numpy().reshape(X_train.shape[0], 2, 1)
    lstm_y_train = y_train.values.reshape(-1, 1)
    lstm_x_test = X_test[['v-2', 'v-1']].to_numpy().reshape(X_test.shape[0], 2, 1)
    lstm_y_test = y_test.values.reshape(-1, 1)

    # Define the LSTM model
    model = Sequential([
        LSTM(50, activation='relu', input_shape=(lstm_x_train.shape[1], lstm_x_train.shape[2])),
        Dropout(0.2),
        Dense(1)
    ])

    # Compile the model
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Train the model
    model.fit(lstm_x_train, lstm_y_train, epochs=120, batch_size=1)
    
    y_pred_test = model.predict(lstm_x_test)
    #test loss (MAE)
    mae = mean_absolute_error(y_test, y_pred_test.reshape(-1))
    #test loss (MAPE)
    mape = MAPE(y_test,y_pred_test.reshape(-1))
    return mae,mape



    
```

```python
def KNN_model(df):
    X = df.drop(['v', 'ISO',"year"], axis=1).values
    y = df['v'].values

    k_values = range(1, 10)
    lowest_mae = float('inf')
    best_k = -1
    mae_scores_list = []

    for k in k_values:
        knn_regressor = KNeighborsRegressor(n_neighbors=k)

        # Compute MAE
        mae_scores = cross_val_score(knn_regressor, X, y, cv=5, scoring='neg_mean_absolute_error')
        avg_mae = np.mean(-mae_scores)
        mae_scores_list.append(avg_mae)

        # Update best k and lowest MAE
        if avg_mae < lowest_mae:
            lowest_mae = avg_mae
            best_k = k

    # Compute MAPE for the best k
    knn_regressor = KNeighborsRegressor(n_neighbors=best_k)
    def mape_scorer(estimator, X, y_true):
        y_pred = estimator.predict(X)
        mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
        return -mape
    mape_score = cross_val_score(knn_regressor, X, y, cv=5, scoring=mape_scorer)
    corresponding_mape = np.mean(-mape_score)

    # Plotting MAE as a function of k
    plt.figure(figsize=(10, 6))
    plt.plot(k_values, mae_scores_list, marker='o')
    plt.title('KNN Regression: MAE for Different Values of k')
    plt.xlabel('Number of Neighbors (k)')
    plt.ylabel('MAE')
    plt.xticks(k_values)
    plt.show()

    return best_k, lowest_mae, corresponding_mape

```

```python
def train_test_splitting(df) :
    #split data
    train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

    # normalize train and test data
    mean_train = train_df[['v-2', 'v-1']].mean()
    std_train = train_df[['v-2', 'v-1']].std()
    train_df[['v-2', 'v-1']] = (train_df[[ 'v-2', 'v-1']] - mean_train) / std_train
    test_df[[ 'v-2', 'v-1']] = (test_df[[ 'v-2', 'v-1']] - mean_train) / std_train
    # seperate features and outcome
    X_train = train_df.drop(['v'], axis=1)
    y_train = train_df['v']
    X_test = test_df.drop(['v'], axis=1)
    y_test = test_df['v']
    return X_train, X_test, y_train, y_test
```
