```python
from pytrends.request import TrendReq
import pandas as pd
import time

# Initialize pytrends
pytrend = TrendReq()

# List of category IDs
category_ids =[356, 255, 288, 5, 233, 1141, 1190, 174, 1299, 1228, 442, 5, 1277, 744, 32, 382, 77, 314, 49, 486, 840, 500, 46, 688, 657, 436, 442, 958, 29, 382, 785, 77, 440, 446, 982, 251, 287, 231, 252, 1327, 1168, 452, 663, 906, 1297, 81, 257, 1143, 268, 1221, 47, 964, 74, 417, 299, 51, 517, 678, 59, 972]

category_names = [
    "Computers & Electronics",  # ID 5
    "Enterprise Technology",    # ID 77
    "Science",                  # ID 174
    "Jobs & Education",         # ID 958
    "Law & Government",         # ID 19
    "Internet & Telecom",       # ID 13
    "Industrial Materials & Equipment",  # ID 287
    "Biotechnology & Pharmaceuticals",   # ID 255
    "Aerospace & Defense",      # ID 356
    "Automotive Industry",      # ID 1190
    "Energy & Utilities",       # ID 233
    "Chemicals Industry",       # ID 288
    "Medical Devices & Equipment",       # ID 251
    "Construction & Maintenance",        # ID 48
    "Health",                   # ID 45
    "Robotics",                 # ID 1141
    "Technology News"           # ID 785
]

# List of countries
countries = [
    "US",  # United States
    "CN",  # China
    "JP",  # Japan
    "DE",  # Germany
    "KR",  # South Korea
    "IN",  # India
    "FR",  # France
    "GB",  # United Kingdom
    "RU",  # Russia
    "BR",  # Brazil
    "CA",  # Canada
    "IT",  # Italy
    "AU",  # Australia
    "ES",  # Spain
    "CH",  # Switzerland
    "NL",  # Netherlands
    "SE",  # Sweden
    "IL",  # Israel
    "FI",  # Finland
    "BE",  # Belgium
    "TW",  # Taiwan
    "SG",  # Singapore
    "AT",  # Austria
    "DK",  # Denmark
    "NO",  # Norway
    "PL",  # Poland
    "TR",  # Turkey
    "PT",  # Portugal
    "IE",  # Ireland
    "CZ"   # Czech Republic
]
# Set timeframe
timeframe = '2005-01-01 2020-12-31'

# Dictionary to hold data frames for each country
country_data_frames = {}

# Loop over each country
for country in countries:
    # Initialize an empty list to collect data frames for this country
    dfs = []
    
    # Fetch the data for each category and add to the list
    for category_id in category_ids:
        attempts = 0
        success = False

        while not success :  # Retry up to 5 times
            try:
                pytrend.build_payload(kw_list=[""], cat=category_id, timeframe=timeframe, geo=country)
                data = pytrend.interest_over_time()
                
                # Drop the 'isPartial' column
                if 'isPartial' in data.columns:
                    data = data.drop(columns=['isPartial'])

                dfs.append(data)
                success = True
            except Exception as e:
                print(f"An error occurred: {e}. Retrying after 60 seconds...")
                time.sleep(1)  # Sleep for 60 seconds
                attempts += 1

    # Concatenate all data frames along the columns
    combined_data = pd.concat(dfs, axis=1)

    # Store the combined data frame in the dictionary
    country_data_frames[country] = combined_data

# Now, country_data_frames contains the data frames for each country
```

```python
for country, df in country_data_frames.items():
    # Check if the number of columns matches the number of category names
        # Rename the columns
        df.columns = category_names
```

```python
# Create a list to hold all DataFrames
all_dataframes = []

# Loop over each country's DataFrame in country_data_frames
for country, df in country_data_frames.items():
    # Add a new column with the country name
    df['Country'] = country  # This adds the country code as a new column

    # Append the DataFrame to the list
    all_dataframes.append(df)

# Concatenate all DataFrames in the list
combined_data = pd.concat(all_dataframes)

# Reset the index of the combined DataFrame
combined_data.reset_index(inplace=True)
combined_data["year"] = combined_data["date"].apply(lambda x : int(str(x).split("-")[0]))
combined_data["month"] = combined_data["date"].apply(lambda x : int(str(x).split("-")[1]))

```

```python
def quartile(month):
    if 1 <= month <= 3:
        return 1  # January to March
    elif 4 <= month <= 6:
        return 2  # April to June
    elif 7 <= month <= 9:
        return 3  # July to September
    else:
        return 4  # October to December
        
```

```python
combined_data["month"] = combined_data["month"].apply(lambda x : quartile(x))
combined_data.rename(columns={'month': 'quartile'}, inplace=True)
combined_data

```

```python
combined_data = combined_data.drop(columns = ["date"])
combined_data_quartiles = combined_data.groupby(['Country', 'year', 'quartile']).mean()
combined_data_quartiles
```

```python
combined_data_quartiles = (combined_data_quartiles  - combined_data_quartiles.mean() )/combined_data_quartiles.std()
combined_data_quartiles = combined_data_quartiles.reset_index()
combined_data_quartiles
```

```python
combined_data_quartiles.to_csv("categories_google_trends.csv",index=False)
```
