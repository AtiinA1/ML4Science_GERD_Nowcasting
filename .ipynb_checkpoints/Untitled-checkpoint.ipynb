```python
from pytrends.request import TrendReq
import pandas as pd

# Initialize pytrends
pytrend = TrendReq()

# List of keywords
keywords = [
    "R&D Expenditure", "Product Development", "Innovation Management",
    "Innovation Strategy", "Government Grants", "Research Funding",
    "Startup Incubation", "Technology Park", "Patent Attorney",
    "Patent Registration", "R&D Jobs", "Collaboration with Industry",
    "Research Grant", "R&D Tax Credit", "VC Investment", "Startup Funding"
]

# List of countries
countries = ['CH']

# Set timeframe
timeframe = '2010-01-01 2020-12-31'

# Split keywords into chunks of 5
keyword_chunks = [keywords[i:i + 5] for i in range(0, len(keywords), 5)]

# Dictionary to hold data frames for each country
country_data_frames = {}

# Loop over each country
for country in countries:
    # Initialize an empty list to collect data frames for this country
    dfs = []
    
    # Fetch the data for each chunk and add to the list
    for chunk in keyword_chunks:
        pytrend.build_payload(kw_list=chunk, timeframe=timeframe, geo=country)
        data = pytrend.interest_over_time()
        
        # Drop the 'isPartial' column
        if 'isPartial' in data.columns:
            data = data.drop(columns=['isPartial'])

        dfs.append(data)

    # Concatenate all data frames along the columns
    combined_data = pd.concat(dfs, axis=1)

    # Store the combined data frame in the dictionary
    country_data_frames[country] = combined_data

# Now, country_data_frames contains the data frames for e
```

```python
country_data_frames["CH"]
switzerland_data = country_data_frames["CH"]

# Define the file name for the CSV
csv_filename = "trends_data_CH.csv"

# Save the DataFrame to CSV
switzerland_data.to_csv(csv_filename, index=True)
```

```python
# Initialize an empty list to collect all data frames with a country column
all_data = []

# Loop over each country and its DataFrame
for country, df in country_data_frames.items():
    # Add a new column to each DataFrame for the country
    df['Country'] = country

    # Append the modified DataFrame to the list
    all_data.append(df)

# Concatenate all DataFrames into a single DataFrame
combined_data = pd.concat(all_data, axis=0)
combined_data["year"] = combined_data.index
```

```python
combined_data["year"] =  combined_data["year"].apply(lambda x : str(x).split("-")[0])
combined_data = combined_data.reset_index(drop=True)
```

```python
combined_data
combined_data_monthly = combined_data
combined_data_monthly = combined_data_monthly.fillna(0)
```

```python
combined_data_yearly =  combined_data_monthly.groupby(['year', 'Country']).mean()
combined_data_yearly = combined_data_yearly.reset_index()
combined_data_yearly
```
