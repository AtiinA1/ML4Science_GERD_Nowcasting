```python
import numpy as np 
import pandas as pd

df_CA= pd.read_csv("./trends_data_CA.csv")
df_CN= pd.read_csv("./trends_data_CN.csv")
df_DE= pd.read_csv("./trends_data_DE.csv")
df_FR= pd.read_csv("./trends_data_FR.csv")
df_GB= pd.read_csv("./trends_data_GB.csv")
df_JP= pd.read_csv("./trends_data_JP.csv")
df_KR= pd.read_csv("./trends_data_KR.csv")
df_US= pd.read_csv("./trends_data_US.csv")
df_CH= pd.read_csv("./trends_data_CH.csv")

countries = ['CAN', 'CHN', 'FRA', 'DEU', 'JPN', 'KOR', 'CHE', 'GBR', 'USA']
country_data_frames = {}
country_data_frames['CAN'] = df_CA
country_data_frames['CHN'] = df_CN
country_data_frames['FRA'] = df_FR
country_data_frames['DEU'] = df_DE
country_data_frames['GBR'] = df_GB
country_data_frames['JPN'] = df_JP
country_data_frames['KOR'] = df_KR
country_data_frames['USA'] = df_US
country_data_frames['CHE'] = df_CH



```

```python
# Initialize an empty list to collect all data frames with a country column
all_data = []

# Loop over each country and its DataFrame
for country, df in country_data_frames.items():
    # Add a new column to each DataFrame for the country
    df['Country'] = country

    # Append the modified DataFrame to the list
    all_data.append(df)

# Concatenate all DataFrames into a single DataFrame
combined_data = pd.concat(all_data, axis=0)
combined_data["year"] =  combined_data["date"].apply(lambda x : str(x).split("-")[0])
combined_data = combined_data.reset_index(drop=True)
combined_data= combined_data.drop(columns="date")
combined_data
```

```python
combined_data_yearly =  combined_data.groupby(['year', 'Country']).mean()
combined_data_yearly = combined_data_yearly.reset_index()
combined_data_yearly = combined_data_yearly.fillna(0)
```
